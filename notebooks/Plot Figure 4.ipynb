```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.insert(1, '../scripts/')

import utilities as utils
```

```python
def find_tree_lineage(i, T):
    p = list(T.predecessors(i))
    curr_node = i
    ancestor_list = [curr_node]
    while p:
        curr_node = p[0]
        ancestor_list.insert(0, curr_node)
        p = list(T.predecessors(curr_node))
    return(ancestor_list)

def find_triplet_structure(triplet, T):
    a, b, c = triplet[0], triplet[1], triplet[2]
    La = find_tree_lineage(a, T.get_tree_topology())
    Lb = find_tree_lineage(b, T.get_tree_topology())
    Lc = find_tree_lineage(c, T.get_tree_topology())
    
    l = 0
    while La[l] == Lb[l] and Lb[l] == Lc[l]:
        l += 1
        
    lca_all = La[l - 1]
    
    g = l
    if La[l] != Lb[l] and La[l] != Lc[l] and Lb[l] != Lc[l]:
        return lca_all, lca_all, None
    
    lca_in = None
    out_group = None
    if La[l] == Lb[l]:
        while La[g] == Lb[g]:
            g += 1
        lca_in = La[g - 1]
        out_group = c
    elif La[l] == Lc[l]:
        while La[g] == Lc[g]:
            g += 1
        lca_in = La[g - 1]
        out_group = b
    else:
        while Lb[g] == Lc[g]:
            g += 1
        lca_in = Lb[g - 1]
        out_group = a
    
    return lca_all, lca_in, out_group
    
def s(u_chars, v_chars):
    k = len(u_chars)
    return sum([u_chars[i] == v_chars[i] and u_chars[i] != 0 and u_chars[i] != -1 for i in range(k)])

def delta_star(p, q, d_star):
    val = 1/(2*p) * np.log((1-q)/q) + 1
    if val < 0:
        delta_star = (1 - q) + q * np.exp(-2 * p)
    elif val >= 0 and val <= d_star:
        delta_star = 2 * np.exp(-p) * np.sqrt(q * (1 - q))
    else:
        delta_star = np.exp(-p * d_star) * (1-q) + q * np.exp(-p * (2 - d_star))
    return delta_star
```

For 100 trees generated from the Asynchronous Simulation, sample 100 triplets from each. Then, for each sampled triplet, we record: 

1) the tree distance between the LCAs of the ingroup and the LCA of the triplet 

2) the difference in the similarity (# shared characters) between the ingroup and the outgroup when they are correctly labeled

3) if the difference in similarity when the outgroup is correclty labeled is less than the threshold t

4) the difference in the similarity (# shared characters) between the ingroup and the outgroup when they are incorrectly labeled

5) if the difference in similarity when the outgroup is incorreclty labeled is more than the threshold t
 

```python
k_cand = 50
lamb = 0.5
num_states = 20
q_dist = dict(zip(range(1, num_states + 1), [1/num_states] * num_states))
depth = 9

dists = []
sims1 = []
wrong1 = 0
sims2 = []
wrong2 = 0

for num in range(100):
    tree = utils.exponential_plus_c_tree_sim(k_cand, q_dist, lamb, depth)
    cm = tree.character_matrix
    thres = 1/2 * cm.shape[1] * lamb * delta_star(lamb, 1/num_states, d_star) * l
    
    for _ in range(100):
        chosen_triplet = np.random.choice(tree.leaves, 3, replace=False)
        lca_all, lca_in, out_group = find_triplet_structure(chosen_triplet, tree)
        lca_all_time = tree.get_attribute(lca_all, "time")
        lca_in_time = tree.get_attribute(lca_in, "time")
        
        dists.append(lca_in_time - lca_all_time)
        in_group = list(set(chosen_triplet) - set([out_group]))
        
        sim_val = s(cm.loc[in_group[0]], cm.loc[in_group[1]]) - max(s(cm.loc[in_group[0]], cm.loc[out_group]), s(cm.loc[in_group[1]], cm.loc[out_group]))
        sims1.append(sim_val)
        if sim_val <= thres:
            wrong1 += 1

        sim_val = s(cm.loc[in_group[0]], cm.loc[out_group]) - max(s(cm.loc[in_group[1]], cm.loc[out_group]), s(cm.loc[in_group[0]], cm.loc[in_group[1]]))
        sims2.append(sim_val)
        if sim_val >= thres:
            wrong2 += 1
```

The following plots figure 4. The ingroup/triplet LCA differences and the ingroup/outgroup similarity differences are plotted for each triplets. Also included are the threshold t, density histograms on both axes, and blue and orange coloring to indicate the cases where the ingroup is identified correctly and incorrectly, respectively.

```python
def scatter_hist(x, y, ax, ax_histx, ax_histy, flag):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    # the scatter plot:
    if flag:
        ax.scatter(x, y, facecolors='none', edgecolors='blue', alpha = 1, s = 10)
    else:
        ax.scatter(x, y, facecolors='none', edgecolors='orange', alpha = 1, s = 10)
    ax.axhline(thres, color='g')

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, alpha = 0.5)
    ax_histy.hist(y, orientation='horizontal', alpha = 0.5)
    

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005


rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a square Figure
fig = plt.figure(figsize=(8, 8))

ax = fig.add_axes(rect_scatter)
ax_histx = fig.add_axes(rect_histx, sharex=ax)
ax_histy = fig.add_axes(rect_histy, sharey=ax)

# use the previously defined function
scatter_hist(dists, sims1, ax, ax_histx, ax_histy, True)
scatter_hist(dists, sims2, ax, ax_histx, ax_histy, False)
print(wrong1/10000)
print(wrong2/10000)
plt.savefig("../Figures/Fig4/triplets.pdf", format='pdf')
plt.show()
```
