This notebook generates the plots for figure 5.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
```

```python
folder = "../results/validate_k/"

regimes = ["percolation_d_triplets", "percolation_d_triplets_20", "percolation_rf", "shared_mutation_rf"]
```

The following generates the heatmaps showing the deviation from the expected 90% reconstruction criteria in the validation of k experiment for the Uniform Simulation.

```python
topology = "complete_binary"
dfs = []

for regime in regimes:
    df = pd.read_csv(f"{folder}{topology}/{regime}.txt", sep= "\t", header = None) 
    
    df = df.drop(columns=[2, 3, 4])
    pivot_df = pd.pivot_table(df, values=5, index=[0], columns=[1])
    
    pivot_df = pd.DataFrame(pivot_df.values - 0.9, index=pivot_df.index, columns=pivot_df.columns)
    
    pivot_df.index = [round(i, 3) for i in pivot_df.index]
    pivot_df.columns = [round(i, 3) for i in pivot_df.columns]
    dfs.append(pivot_df)

    divnorm=colors.DivergingNorm(vmin=pivot_df.min().min(), vcenter=0., vmax=pivot_df.max().max())

    plt.pcolor(pivot_df, cmap = 'coolwarm', norm=divnorm)
    plt.colorbar()
    plt.yticks(np.arange(0.5, len(pivot_df.index), 1), pivot_df.index)
    plt.xticks(np.arange(0.5, len(pivot_df.columns), 1), pivot_df.columns)
    ax = plt.axes()
    ax.set(facecolor = "moccasin")

    plt.savefig(f"../Figures/Fig5/validate_k_{topology}_{regime}.eps", format='eps')
    plt.show()
```

The following generates the heatmaps showing the deviation from the expected 90% reconstruction criteria in the validation of k experiment for the Asynchronous Simulation.

```python
topology = "exponential_plus_c"

dfs = []
std_dfs = []

for regime in regimes:
    df = pd.read_csv(f"{folder}{topology}/{regime}.txt", sep= "\t", header = None) 

    df = df.drop(columns=[2, 3, 4])
    mean_df = df.set_index([0, 1]).groupby([0, 1]).mean()
    pivot_df = pd.pivot_table(mean_df, values=5, index=[0], columns=[1])
    
    pivot_df = pd.DataFrame(pivot_df.values - 0.9, index=pivot_df.index, columns=pivot_df.columns)
    
    pivot_df.index = [round(i, 3) for i in pivot_df.index]
    pivot_df.columns = [round(i, 3) for i in pivot_df.columns]

    std_df = df.set_index([0, 1]).groupby([0, 1]).std()
    std_pivot_df = pd.pivot_table(std_df, values=5, index=[0], columns=[1])

    divnorm=colors.DivergingNorm(vmin=pivot_df.min().min(), vcenter=0., vmax=pivot_df.max().max())

    plt.pcolor(pivot_df, cmap = 'coolwarm', norm=divnorm)
    plt.colorbar()
    plt.yticks(np.arange(0.5, len(pivot_df.index), 1), pivot_df.index)
    plt.xticks(np.arange(0.5, len(pivot_df.columns), 1), pivot_df.columns)
    ax = plt.axes()
    ax.set(facecolor = "moccasin")

    for y in range(pivot_df.shape[0]):
        for x in range(pivot_df.shape[1]):
            plt.text(x + 0.5, y + 0.5, '%.2f' % round(std_pivot_df.iloc[y, x], 2),
                     horizontalalignment='center',
                     verticalalignment='center',
                     color = 'black'
                     )
            
    plt.savefig(f"../Figures/Fig5/validate_k_{topology}_{regime}.eps", format='eps')
    
    dfs.append(pivot_df)
    std_dfs.append(std_pivot_df)
    
    plt.show()
```
